**8. Clustering Analysis**

```python
!pip install openpyxl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

import warnings
warnings.filterwarnings('ignore')

# Load dataset using pd.read_excel for .xlsx files
EWA=pd.read_excel('/content/EastWestAirlines.xlsx',sheet_name='data')
EWA.head()
```

```python
# rename the columns
EWA.rename(columns={'ID#':'cum_ID','cc1_miles':'FFcc_miles','cc2_miles':'Rcc_miles','cc3_miles':'SBcc_miles','Award?':'Award'},
inplace=True)
```

```python
EWA.info()
```

```python
# Droping cum_id column
EWA_1=EWA.drop('cum_ID',axis=1)
```

```python
sns.pairplot(EWA_1)
```

```python
# Standardising the data as euclidean distance is influenced by scaling
std=StandardScaler()
EWA_std=std.fit_transform(EWA_1)
EWA_std=pd.DataFrame(EWA_std,columns=EWA_1.columns)
```

**Implementing clustering Algorithms**

K means

```python
wcss=[]
k=[]
for i in range(1,11):
    kmean=KMeans(n_clusters=i,n_init=10)
    kmean.fit(EWA_std)
    wcss.append(kmean.inertia_)
    k.append(i)
```

```python
# Scree plot or Elbow plot
x=np.array(k)
y=np.array(wcss)
plt.plot(x,y)
plt.show()
```

```python
# from the elbow graph it is clear to get 4 clusters
km=KMeans(n_clusters=4,n_init=10)
km.fit(EWA_std)
```

```python
labels=km.labels_
```

```python
ewa_k=EWA_1.copy()
ewa_k['label']=labels
```

```python
ewa_k.label.value_counts()
```

```python
# Evaluation and Performance Metrics:
ss=silhouette_score(EWA_std,labels)
ss
```

```python
ewa_k.groupby(['label','Award']).agg('mean')
```

DBSCAN

```python
# first calculate best epsilon value using K-distance graph
from sklearn.neighbors import NearestNeighbors # Import the NearestNeighbors class
neigh= NearestNeighbors(n_neighbors=4)
nbrs=neigh.fit(EWA_std)
distances,indices=nbrs.kneighbors(EWA_std)
distances = np.sort(distances, axis=0)
```

```python
# Plotting K-distance Graph
distance = distances[:,1]
plt.figure(figsize=(20,10))
plt.plot(distance)
plt.title('K-distance Graph',fontsize=20)
plt.xlabel('Data Points sorted by distance',fontsize=14)
plt.ylabel('Epsilon',fontsize=14)
plt.grid()
plt.show()
```

```python
# take epsilon value = 1.5
dbscan_opt=DBSCAN(eps=1.5,min_samples=6) # i checked the min_sampels from 4 ,5,6
dbscan_opt.fit(EWA_std)
```

```python
ewa=EWA_1.copy()
ewa['DBSCAN_opt_labels']=dbscan_opt.labels_
ewa['DBSCAN_opt_labels'].value_counts()
```

```python
# Evaluation and Performance Metrics:
s=silhouette_score(EWA_std,dbscan_opt.labels_)
s
```

```python
ewa.groupby(['DBSCAN_opt_labels','Award']).agg('mean')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

import warnings
warnings.filterwarnings('ignore')

# Load dataset using pd.read_excel for .xlsx files
EWA=pd.read_excel('/content/EastWestAirlines.xlsx',sheet_name='data')
EWA.head()

# rename the columns
EWA.rename(columns={'ID#':'cum_ID','cc1_miles':'FFcc_miles','cc2_miles':'Rcc_miles','cc3_miles':'SBcc_miles','Award?':'Award'},
inplace=True)

EWA.info()

# Droping cum_id column
EWA_1=EWA.drop('cum_ID',axis=1)

sns.pairplot(EWA_1)

# Standardising the data as euclidean distance is influenced by scaling
std=StandardScaler()
EWA_std=std.fit_transform(EWA_1)
EWA_std=pd.DataFrame(EWA_std,columns=EWA_1.columns)

import scipy.cluster.hierarchy as sch # Import the module and assign it to the alias 'sch'

dm = sch.linkage(EWA_std, method='complete', metric='mahalanobis') #Now sch can be used here
```

```python
dendo=sch.dendrogram(dm)
```

```python
hc = AgglomerativeClustering(n_clusters=4,metric='mahalanobis',linkage='complete')
hc.fit(EWA_std)
```

```python
EWA_hc=EWA_1.copy()
```

```python
EWA_hc['labels']=hc.labels_
```

```python
EWA_hc.labels.value_counts()
```

```python
EWA_hc.groupby(['labels','Award']).agg('mean')
```
